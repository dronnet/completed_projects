```python
#код сочиняет другой код для SQL, для автоматического кодирования признаков от 1 до....
#Значения None кодируются 0
def coder_sql(column_name):
    col = column_name.columns.values[0]
    print(', CASE')
    count = 1
    for j in column_name[col]:
    
        if j != None:
            print(f"    WHEN {col} = '{j}' THEN {count}")
            count += 1
    print('    ELSE 0')
    print(f'END c_{col}')
```

По условию требуется использовать SQL запрос, что и сделаю.

```python
query = '''
SELECT DATE(DATE_TRUNC('month', collision_date)), count(*)
FROM collisions
GROUP BY DATE_TRUNC('month', collision_date)
--LIMIT 5
; 
'''
crash_month = pd.read_sql_query(query, con=engine)
crash_month.head()
#crash_month.sort_values(by='count')
```

```python
#количество происшествий по месяцам каждого года
sns.set(rc={"figure.figsize":(12, 16)})
sns.barplot(data=crash_month, y="date", x="count");
```

Есть небольшая зависимость от месяца. Первый и второй месяцы каждого года наблюдается снижение количества аварий, но небольшое, около 10-12%. 

    В остальные месяцы какой либо зависимости не просматривается

Вызывает вопросы 2012 год. Заказчик желает использовать данные за этот год, как самые свежие. 

Начиная с 6-го месяца этого 2012 года, количество аварий сильно меньше чем за первые месяцы, примерно в 12 раз. 

А следующие месяцы 7,8,9 и т.д. вообще имеют мало данных. 

Получается что брать данные за 2012 год не выгодно, так как за большую половину 2012 года данных слишком мало.

Лучше будет взять данные за предыдущие года, например с 2009-го по 2011-й включительно.

3. Есть ли связь между типом кузова автомобиля и аварийностью?
 - создать запрос sql (связать таблицы `collisions` и `vehicles`) 
 - построить три графика, за 2009,2010,2011 -  количество аварий от вида кузова
 - краткий вывод

```python
query = '''
WITH veh_type AS
(SELECT case_id, vehicle_type
FROM vehicles
),

col_date AS
(SELECT case_id, DATE(collision_date)
FROM collisions
WHERE collision_date BETWEEN '2009-01-01' AND '2011-12-31'
),

join_tabl AS
(SELECT *
FROM veh_type  JOIN col_date ON col_date.case_id=veh_type.case_id
ORDER BY col_date.case_id)

SELECT vehicle_type, COUNT(*)
FROM join_tabl
GROUP BY vehicle_type
; 
'''

task_3 = pd.read_sql_query(query, con=engine) 
```

## Создайте модель для оценки водительского риска

в задании от заказчика требуется учесть возраст автомобиля. Он есть только в таблице `vehicles`.
Следовательно необходимо использовать  все три таблицы `vehicles` `collisions` и `parties`

У меня уже есть все таблицы с уже кодированными данными, которые занимают немного памяти и можно использовать их для запроса.

Но в задании требуется использовать sql запрос, так и поступлю.

```python
query = '''
WITH part AS
(SELECT *
FROM parties
WHERE party_type = 'car'
ORDER BY case_id, party_number
--LIMIT 20
), 

col AS
(SELECT *
FROM collisions
WHERE collision_date BETWEEN '2012-01-01' AND '2012-12-31' AND collision_damage NOT IN ('scratch')
--LIMIT 20
),

veh AS
(SELECT case_id, party_number, vehicle_age
FROM vehicles
)

SELECT --id -- номер для модели не нужен
        --, case_id -- номер для модели не нужен
        --, party_number -- номер участника для модели не нужен
        --, party_type -- уже и так выбрали тип участника "CAR", столбец заполнен только этим значением
        at_fault -- ЦЕЛЕВОЙ ПРИЗНАК !!
        --, insurance_premium -- суммма страховки - клиент получает её после аварии, не влияет на аварию
        , party_sobriety -- "трезвость участника" однозначно влияет
        , party_drug_physical -- "состояние участника от лекарств" тоже может влиять, особенно если лекарства меняют сознание
        , cellphone_in_use -- "телефон в авто" по идее не должно влиять, но всё может быть
        --, case_id
        --, county_city_location
        --, county_location -- Теоретически район может влиять на аварийность, но маловероятно, не буду использовать
        , distance
        --, direction -- "Направление движения", не думаю что это влияет на аварийность
        , intersection -- "перекрёсток" аварии часто происходят на перекрёстках, пока оставлю, но вообще человек в любом случае ездит через перекрёстки
        , weather_1 -- "погода" явно влияет
        , location_type -- "тип дороги" - наверняка имеет влияние
        --, collision_damage --повреждения уже после аварии и не могут быть использованы для предсказания
        -- , party_count --"число участников" никак не влияет на предсказания
        --, primary_collision_factor --"Основной фактор аварии" - невозможно предугадать
        --, pcf_violation_category --"Категория нарушения" так же невозможно предугадать
        --, type_of_collision --Это уже после аварии
        --, motor_vehicle_involved_with -- (доп участники дтп) До поездки нельзя предугадать кто поучавствует в аварии
        , road_surface --"Состояние дороги" должно оказывать влияние
        , road_condition_1 --"Дорожное состояние" однозначно влияет.
        , lighting --"Освещение" влияет
        , control_device -- Не понял что это такое, пока оставлю
        , EXTRACT('dow' FROM collision_date) AS day --Думаю именно дата не сильно влияет, может влиять день (понедельник, вторник и т.д.)
        -- перевёл дату в код дня, от 0 до 6
        , EXTRACT('hour' FROM collision_time) AS hour -- Оставлю только час аварии, возможно время аварии влияет на аварийность
        , vehicle_age
FROM part JOIN col ON part.case_id=col.case_id JOIN veh ON part.case_id=veh.case_id AND part.party_number=veh.party_number
--LIMIT 5
; 
'''
tabl_all = pd.read_sql_query(query, con=engine) 
tabl_all.head(5)
```

При `LEFT JOIN` таблицы `vehicles` получается таблица из почти 200 000 строк. Если использовать `INNER JOIN`, то выходная таблица получается небольшой около 55 000 строк, при этом в колонке `vehicle_age` немного пропусков.
