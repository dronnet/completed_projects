- датасет имеет 26496 строк и все строки идут последовательно, с линейным возрастанием времени, монотонно.

  - не наблюдается неадекватных выбросов
  - гистограмма имеет вид нормального распределения. Только нет "хвоста" для отрицательных значений
- согласно преобразованию Фурье:
  - Имеется большая постоянная составляющая около 14 единиц, то есть среднее число поездок за всё время 14. Это же число подтверждается расчётом среднего значения.
  - Эта постоянная составляющая намного выше чем уровень гармоник и явно выше чем их сумма, что логично, так как число поездок не может быть отрицательно. И следовательно гармоники должны быть намного ниже чем постоянная составляющая
  - самые сильные гармоники имеют частоту 0,041 1/час и 0,082 1/час что соответствует периодам 1/0,041 = 24ч и 1/0,082 = 12 часов. Это тоже логично так как такси более активно днём (раз в 24 часа), и более активно утром и вечером (два раза в 24 часа).
  - так же имеются более высокочастотные гармоники, что показывает наличие и других периодических изменений количество поездок во времени.
  - Так как спектр имеет яркие гармоники и не похож на шум, это даёт возможность прогнозировать значения. Если бы гармоник не было и ДПФ представляло собой спектр шума то прогнозы были бы невозможны.
  - выше частоты 1,0 гармоник практически нет или они слабые. Значит для устранения лишнего шума можно применить фильтр нижних частот с частотой среза примерно 1,0...1,2. В принципе ресемплирование по 1 часу примерно это и делает.
  
- согласно декомпозиции:
  - имеется постоянно растущий тренд. То есть количество поездок неуклонно увеличивается
  - Сезонная составляющая одинакова за любой период и любой месяц. Имеет периодичность 24 часа
  - В ночные часы кол-во поездок падает, потом увеличивается в утренние часы (9:00), днем (16:00) и ночью около 24:00
  - Так же на сезонной составляющей видна общая периодичность в 24 часа.
  - Нет периодичности от дня, недели или месяца.
  

### создание признаков

Так как не видно зависимости от года (во всяком случае данные только за пол года), нет зависимости от месяца и дня, то добавлю только признак часа

```python
#функция для создания признаков
def make_features(data, max_lag, rolling_mean_size):
    #data['year'] = data.index.year
    #data['month'] = data.index.month
    #data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
#создание копии данных, чтобы не перезапускать весь код сверху
data_copy = data.copy()
```

```python
make_features(data_copy, 48, 2)
```

```python
#проверка
data_copy.head(10)
```

```python
train, test = train_test_split(data_copy, shuffle=False, test_size=0.1)
```

```python
#удаление строк с пустыми значениями в ячейках
train = train.dropna()
```

```python
train_x = train.drop('num_orders', axis=1)
train_y = train['num_orders']
```

## Обучение

###### подготовка данных для обучения с помощью TimeSeriesSplit

```python
#отрезаем 1/10 часть от тренировочных данных для валидации гридсерча
tscv = TimeSeriesSplit(n_splits=2, test_size=int(len(train_x)/10))
```

```python
#проверка правильности разделения выборки на тренировочную и валидационную
for i, (train_index, test_index) in enumerate(tscv.split(train_x)):
    print(f"Fold {i}:")
    print(f"  Train: index_min={train_index.min()}")
    print(f"  Train: index_max={train_index.max()}")
    print(f"  Test:  index_min={test_index.min()}")
    print(f"  Test:  index_max={test_index.max()}")
```

```python
len(train_x)
```

#### LinearRegression

```python
params = {} 
```

```python
lr = LinearRegression()
```

```python
clf = GridSearchCV(lr,param_grid=params , cv=tscv, scoring='neg_mean_squared_error', n_jobs=-1)
```
